**Importing required libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

**Loading Dataset**

```python
cols=['Sepal len','Sepal wid','Petal len','Petal wid','Class_labels']
iris=pd.read_csv("/content/sample_data/iris.data",names=cols)
iris.head()
```

**Information of the dataset**

```python
iris.describe()
```

**Pair** **Plot**

```python
sb.pairplot(iris,hue='Class_labels')
```

**Training and Testing Data**

```python
data=iris.values
x=data[:,0:4]
y=data[:,4]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(len(y_train))
```

**Model 1: Support Vector Machine(SVM)**

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(x_train,y_train)
```

```python
pre1=model_svc.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pre1)*100)
for i in range(len(pre1)):
    print(y_test[i],pre1[i])
```

**Model 2: Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(x_train,y_train)
```

```python
pre2=model_LR.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,pre2)*100)
```
